1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#First connect to the driver

driver= webdriver.Chrome(r"chromedriver_win32.exe")

driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

driver.maximize_window()
```

```python
rank_mv = []
name_mv = []
artist_mv = []
upload_date_mv = []
views_mv = []
```

```python
rank = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for r in rank:
    rank_mv.append(r.text[:-1])

name = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for n in name:
    n1 = n.text
    n2 = re.findall(r'[a-zA-Z]+', n1)
    n3 = ''
    for i in n2:
        n3 = n3 + ' ' + i
    name_mv.append(n3)

    
artist = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for a in artist:
    artist_mv.append(a.text)

views = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for v in views:
    views_mv.append(v.text)
    
upload_date = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for ud in upload_date:
    upload_date_mv.append(ud.text)
```

```python
print(len(rank_mv))
print(len(name_mv))
print(len(artist_mv))
print(len(views_mv))
print(len(upload_date_mv))
```

```python
# Creating a DataFrame
most_viewed_df = pd.DataFrame({'Rank': rank_mv, 
                               'Video Name': name_mv, 
                               'Artist': artist_mv, 
                               'Upload Date': upload_date_mv, 
                               'Views(in Billions)': views_mv})
print('-------------------------Most viewed videos on YouTube from Wikipedia-------------------------')
most_viewed_df.set_index(['Rank'], inplace = True)
most_viewed_df
```

4. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
#First connect to the driver

driver= webdriver.Chrome(r"chromedriver_win32.exe")
```

```python
driver.get('https://github.com/')

driver.maximize_window()
```

```python
move = driver.find_elements(By.XPATH, '//button[@class="HeaderMenu-link border-0 width-full width-lg-auto px-0 px-lg-2 py-3 py-lg-2 no-wrap d-flex flex-items-center flex-justify-between js-details-target"]')
for m in move:
    if m.text == 'Open Source':
        open_source = m
action = webdriver.ActionChains(driver)
action.move_to_element(open_source)
action.perform()
```

```python
trending = driver.find_elements(By.XPATH, '//a[@class="HeaderMenu-dropdown-link lh-condensed d-block no-underline position-relative py-2 Link--secondary"]')
for t in trending:
    t1 = t.get_attribute('href')
    if t1 == 'https://github.com/trending':
        t.click()
```

```python
repository_title = []
repository_description = []
contributors_count = []
languages_used = []

# Trending repository title
title = driver.find_elements(By.XPATH, '//h2[@class="h3 lh-condensed"]/a')
for rt in title:
    rt1 = rt.text
    rt2 = re.findall(r'/.+', rt1)
    rt3 = re.sub(r"[\([{})\]]", "", str(rt2))
    repository_title.append(rt3.replace("'/ ",''))

# Description of the repository
description = driver.find_elements(By.XPATH, '//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for d in description:
    repository_description.append(d.text)

# Contributors count
contributor1 =driver.find_elements(By.XPATH, '//*[@class="f6 color-fg-muted mt-2"]')
for c1 in contributor1:
    c2 = c1.text.split('\n')[1]
    c3 = c2.replace(' stars today', '')
    contributors_count.append(c3)

# Programming languages used
l1 = []  
languages = driver.find_elements(By.XPATH, '//*[@class="f6 color-fg-muted mt-2"]')
for l2 in languages:
    l3 = (l2.text.split('\n')[0]).replace(' Built by','')
    l4 = re.findall(r'[a-zA-Z]+',l3)
    l5 = re.sub(r"[\([{})\]]", "", str(l4))
    l1.append(l5)
for lu in range(len(l1)):
    languages_used.append(l1[lu])
```

```python
print(len(repository_title))
print(len(repository_description))
print(len(contributors_count))
print(len(languages_used))
```

```python
# Creating a DataFrame
git_rep_df =pd.DataFrame({'Repository Title':repository_title[0:23], 'Description':repository_description[0:23], 
                            'Contributors Count':contributors_count[0:23], 'Languages used':languages_used[0:23]})
print('-'*25,'Trending repositories on Github.com','-'*25)
git_rep_df
```

```python
driver.get('https://www.imdb.com/list/ls095964455/')

driver.maximize_window()
```

```python
series_name = []
year_span = []
series_genre = []
runtime = []
series_ratings = []
series_votes = []


# Series Name
series_n = driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]/a')
for sn in series_n:
    series_name.append(sn.text)
    
# Series span years
years1 = driver.find_elements(By.XPATH, '//span[@class="lister-item-year text-muted unbold"]')
for y1 in years1:
    year_span.append(y1.text)
    
# Series genre
genre = driver.find_elements(By.XPATH, '//span[@class="genre"]')
for g in genre:
    series_genre.append(g.text)
    
# Series Run time
run_t = driver.find_elements(By.XPATH, '//span[@class="runtime"]')
for rt in run_t:
    runtime.append(rt.text)
    
# Series ratings
s_rating = driver.find_elements(By.XPATH, '//div[@class="ipl-rating-star small"]')
for r in s_rating:
    series_ratings.append(r.text)
    
# Series votes
votes = driver.find_elements(By.XPATH, '//p[@class="text-muted text-small"][3]/span[2]')
for v in votes:
    series_votes.append(v.text)
```

```python
print(len(series_name), len(year_span), len(series_genre), len(runtime), len(series_ratings),len(series_votes))
```

```python
# Creating a DataFrame
imdb_top_series_df = pd.DataFrame({'Name':series_name,'Years Span':year_span,'Genre':series_genre, 
                                   'Run Time':runtime,'Ratings':series_ratings, 'Votes':series_votes})
print('-'*25, 'Most watched tv series of all time from imdb.com', '-'*25)
imdb_top_series_df
```
