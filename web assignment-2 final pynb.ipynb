```python
#WEB SCRAPING – ASSIGNMENT 2


#Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You 
# have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 
# jobs data.
# This task will be done in following steps:
# 1. First get the webpage https://www.naukri.com/
# 2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the 
# location” field.
# 3. Then click the searchbutton.
# 4. Then scrape the data for the first 10 jobs results you get.
# 5. Finally create a dataframe of the scraped data.
# Note: All of the above steps have to be done in code. No step is to be done manually.
```

```python
#import all the required dictionary
import selenium as sm
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#now install the crome driver and connect to the driver which is available in this folder of my laptop
driver = webdriver.Chrome(r"C:\Users\Pavan Krishna\Downloads\chromedriver_win32 (1).exe")
```

```python
# Now opening the naukri.com website in a new window
driver.get("https://www.naukri.com/")
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys("Data Analyst")
```

```python
#NOW SEARCH FOR the location by automated test software
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Bengaluru")

search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
#now search for all the information which is necessary
#searching for job title
T=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in T[0:10]:
    T=i.text
    job_title.append(T)

#searching for company name   
C=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in C[0:10]:
    C=i.text
    company_name.append(C)

#searching for job location
L=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in L[0:10]:
    L=i.text
    job_location.append(L)

#searching for job experience
E=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in E[0:10]:
    E=i.text
    experience_required.append(E)
```

```python
#at last creat the dataframe
df = pd.DataFrame({' job_title ':job_title,' Job_location ':job_location,' company_name ':company_name,' experience_required ':experience_required})
```

```python
# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
# The above task will be done in following steps:
# 1. First get the webpagehttps://www.azquotes.com/
# 2. Click on TopQuotes
# 3. Than scrap a) Quote b) Author c) Type Of Quotes
```

```python
driver = webdriver.Chrome(r"C:\Users\Pavan Krishna\Downloads\chromedriver_win32 (1).exe")
```

```python
#url search
url="https://www.azquotes.com/"
driver.get(url)
```

```python
#webscrapping of types of quotes
for page in range(0, 10):
    q=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in q:
        q=i.text
        Quotes.append(q)
    next_page=driver.find_elements(By.XPATH,'//li[@class="next"]')
    
    
print(len(Quotes))
print(Quotes)
```

```python
#webscrapping of types of authors
for page in range(0, 10):
    a=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in a:
        a=i.text
        Author.append(a)
    next_page=driver.find_elements(By.XPATH,'//li[@class="next"]')
    
    
print(len(Author))
print(Author)
```

```python
#webscrapping of types of quotes
for page in range(0, 10):
    t=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in t:
        t=i.text
        Type_Of_Quotes.append(t)
    next_page=driver.find_elements(By.XPATH,'//li[@class="next"]')
    
    
print(len(Type_Of_Quotes))
print(Type_Of_Quotes)
```
